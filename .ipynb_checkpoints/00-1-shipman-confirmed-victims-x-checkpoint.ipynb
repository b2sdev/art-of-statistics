```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("./data/00-1-shipman-confirmed-victims-x.csv")
```

```python
g = sns.jointplot(data=df, x="fractionalDeathYear", y="Age", hue="gender2", xlim=(1974,1998), ylim=(39,95))
g.set_axis_labels('Year', 'Age of victim', fontsize=16)
g.plot(sns.scatterplot, sns.histplot)
```
